## Завдання 1

Діедральна група $D_{20}$ має ряд нормальних підгруп: 

Для кожної нормальної підгрупи $N$ побудуйте ендоморфізм $\phi : D_{20} \rightarrow D_{20}$ такий, що $Ker \phi = N$. Задайте його через `PermutationGroupMorphism` і використайте метод `.kernel` як підтвердження. 

___Зауваження___: ядро має бути не __ізоморфне__ підгрупі, а __рівне__ їй.



```python
D20 = DihedralGroup(20)
normal_subgroups = [N for N in D20.normal_subgroups()]
morphisms = []
for N in normal_subgroups:
    quotient, projection = D20.quotient(N)
    gens = D20.gens()
    images = [projection(g) for g in gens]
    morphism = PermutationGroupMorphism(D20, quotient, images)
    assert morphism.kernel() == N
    morphisms.append((N, morphism))
morphisms
```

## Завдання 4

1. Знайдіть прості графи з найменшою к-тю вершин, в яких групи автоморфізмів ізоморфні $A_2$ і $A_3$ відповідно. 
2. (додаткова) Зробіть те саме для $A_4$ i $A_5$. 



```python
from sage.groups.perm_gps.permgroup_named import AlternatingGroup
from sage.graphs.graph_generators import graphs
def find_graph_with_automorphism_group(target_group):
    for n in range(1, 10):
        for G in graphs(n):
            if G.automorphism_group().is_isomorphic(target_group):
                return G
    return None
A2 = AlternatingGroup(2)
A3 = AlternatingGroup(3)
A4 = AlternatingGroup(4)
A5 = AlternatingGroup(5)
graph_A2 = find_graph_with_automorphism_group(A2)
graph_A3 = find_graph_with_automorphism_group(A3)
graph_A4 = find_graph_with_automorphism_group(A4)
graph_A5 = find_graph_with_automorphism_group(A5)
print("Граф для A_2:", graph_A2)
print("Граф для A_3:", graph_A3)
print("Граф для A_4:", graph_A4)
print("Граф для A_5:", graph_A5)
if graph_A2:
    graph_A2.show(title="Граф для A_2")
if graph_A3:
    graph_A3.show(title="Граф для A_3")
if graph_A4:
    graph_A4.show(title="Граф для A_4")
if graph_A5:
    graph_A5.show(title="Граф для A_5")
```
