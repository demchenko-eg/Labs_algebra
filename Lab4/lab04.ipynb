## Завдання 1. 

__Знайдіть усі незвідні системи твірних групи $S_4$.__

```python
G = SymmetricGroup(4)
elements = G.list()
irreducible_generating_sets = []

for g1 in elements:
    for g2 in elements:
        if G.subgroup([g1, g2]) == G:
            irreducible_generating_sets.append((g1, g2))

irreducible_generating_sets
```

## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
S4 = SymmetricGroup(4)
A4 = AlternatingGroup(4)
S5 = SymmetricGroup(5)
A5 = AlternatingGroup(5)

subgroups_S4 = S4.subgroups()
subgroups_A4 = A4.subgroups()
subgroups_S5 = S5.subgroups()
subgroups_A5 = A5.subgroups()

with open("subgroups.txt", "w") as f:
    f.write("Subgroups of S4:\n")
    for subgroup in subgroups_S4:
        f.write(str(subgroup) + "\n")
    f.write("\nSubgroups of A4:\n")
    for subgroup in subgroups_A4:
        f.write(str(subgroup) + "\n")
    f.write("\nSubgroups of S5:\n")
    for subgroup in subgroups_S5:
        f.write(str(subgroup) + "\n")
    f.write("\nSubgroups of A5:\n")
    for subgroup in subgroups_A5:
        f.write(str(subgroup) + "\n")
```

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
import random

n = random.randint(1, 100)
G = SymmetricGroup(n)

g1 = G.random_element()
g2 = G.random_element()

is_g1_even = g1.sign() == 1
is_g2_even = g2.sign() == 1

generated_group = G.subgroup([g1, g2])
order_generated_group = generated_group.order()

A_n = AlternatingGroup(n)
is_generated_group_equal_A_n = generated_group.is_isomorphic(A_n)
is_generated_group_equal_S_n = generated_group.is_isomorphic(G)

print("Підстановка g1:", g1)
print("Чи є g1 парною?", is_g1_even)

print("Підстановка g2:", g2)
print("Чи є g2 парною?", is_g2_even)

print("Порядок групи, породженої g1 та g2:", order_generated_group)
print("Чи дорівнює породжена група A_n?", is_generated_group_equal_A_n)
print("Чи дорівнює породжена група S_n?", is_generated_group_equal_S_n)
```
